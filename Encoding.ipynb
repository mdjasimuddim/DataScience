# **loop**

```python

for col in df8.columns:
    if is_numeric_dtype(df8[col]):         # If each column of df8 is of numeric type, then continue
        continue
    else:
        one1 = pd.get_dummies(df8[col], drop_first = True)    #Getting dummy varibles from that non-numeric column, dropping 1st column among the dummy columns and putting the other dummies into 'one1' variable.
        df8.drop(df8[[col]], axis = 1, inplace = True)        # Deleting the older column of df8 and,
        df8 = pd.concat([df8, one1], axis = 1)                #then add the excess columns of df8 with the remaining dummy columns kept inside one1.
```

# **Ordinal Encoder**

```python
df9.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
order0 = df9.Area.unique()          #'order0' variable contains the unique values of 'sex' column and we can set the order.
order0                             #order0 = ['Dhaka',  'ctg','Rangpur']
```

```python
ordinal = OrdinalEncoder(categories=[order0])      # Creating an object for OrdinalEncoder.
```

```python
ordinal   #by default Dhaka is first, then ctg and last Rangpur
```

```python
encoded = ordinal.fit_transform(df9[['Area']])    #Ordinal encoding is done on 'Area' column and kept inside 'encoded' variable.
```

```python
encoded
```

```python
df9.Area = encoded
```

```python
df9.head()
```

```python
#loop
for col in df10.columns:
    if is_numeric_dtype(df10[col]):
        continue
    else:
        order = df10[col].unique()                            # Setting the order of non-numeric columns in df10.
        df10[col] = OrdinalEncoder(categories=[order]).fit_transform(df10[[col]])    # Encode that column of df10.
```
