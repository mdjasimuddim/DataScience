# MinMax Scaler

```python
from sklearn.preprocessing import MinMaxScaler #import MinMaxScaler from sklearn.preprocessing

min_max_scale = MinMaxScaler()
df['Marketing Spend'] = min_max_scale.fit_transform(df[['Marketing Spend']])
```

```python
df.head()
```

## Max Absolute Scaler

```python
from sklearn.preprocessing import MaxAbsScaler   #import MaxAbsScaler from sklearn.preprocessing

max_abs_scale = MaxAbsScaler()
df['Marketing Spend'] = max_abs_scale.fit_transform(df[['Marketing Spend']])
```

```python
df.head()
```

## Robust Scaler

```python
from sklearn.preprocessing import RobustScaler  #import RobustScaler from sklearn.preprocessing|
```

```python
robust_Scaler = RobustScaler()
```

```python
df['Marketing Spend'] = robust_Scaler.fit_transform(df[['Marketing Spend']])
```
